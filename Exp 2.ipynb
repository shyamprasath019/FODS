```python
#Handling Null Values
```

```python
import pandas as pd
```

```python
df=pd.read_csv("housing_dataset.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
price_Med=df["Price"].median()
price_Med
```

```python
df["Price"]=df["Price"].fillna(value=price_Med)
```

```python
df["Bathrooms"]=df["Bathrooms"].fillna(df["Bathrooms"].mean())
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
#Handling Outliers
```

```python
Q1=df['Price'].quantile(0.25)
Q3=df['Price'].quantile(0.75)
IQR=Q3-Q1
```

```python
lb=Q1-1.5*IQR
up=Q3+1.5*IQR
```

```python
outlier = df[ (df['Price']<lb) | (df['Price']>up) ]
```

```python
outlier["Price"]
```

```python
df=df.drop(outlier.index)
```

```python
df.head(10)
```

```python
normalised=df.copy()
column='Sqft_Living'
```

```python
normalised[column]=(normalised[column]-normalised[column].min())/(normalised[column].max()-normalised[column].min())
```
