```python
import pandas as pd
```

```python
df = pd.ExcelFile(r"C:\Users\shyam\Downloads\payroll_dataset.xlsx")
```

```python
employees_df = pd.read_excel(df, sheet_name='Employees')
departments_df = pd.read_excel(df, sheet_name='Departments')
```

```python
merged_df = pd.merge(employees_df, departments_df, on='DepartmentID')
merged_df.head()
```

```python
merged_df.sum().isnull()
```

```python
cleaned_df = merged_df.drop_duplicates()
rows_before = merged_df.shape[0]
rows_after = cleaned_df.shape[0]
```

```python
rows_before,rows_after
```

```python
Q1 = cleaned_df['Salary'].quantile(0.25)
Q3 = cleaned_df['Salary'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers_removed_df = cleaned_df[(cleaned_df['Salary'] >= lower_bound) & (cleaned_df['Salary'] <= upper_bound)]
```

```python
rows_before2 = cleaned_df.shape[0]
rows_after2 = outliers_removed_df.shape[0]
```

```python
lower_bound, upper_bound, rows_before2, rows_after2
```

```python
encoded_df = pd.get_dummies(cleaned_df, columns=['Location'])
encoded_df
```

```python
min_salary = cleaned_df['Salary'].min()
max_salary = cleaned_df['Salary'].max()
cleaned_df['Salary_Normalized'] = (cleaned_df['Salary'] - min_salary) / (max_salary - min_salary)
```
